**Loading Dataset**

```python
import pandas as pd
import numpy as np
from sklearn.neighbors import NearestNeighbors
from sklearn.preprocessing import StandardScaler
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.pipeline import Pipeline
```

```python
file_path = '/content/recipe_final (1).csv'
df = pd.read_csv(file_path)
df.head()
```

```python
# preprcoss ingredients
vector = TfidfVectorizer()
X = vector.fit_transform(df['ingredients_list'])
```

```python
df
```

```python
# normalizing numerical features
scaler = StandardScaler()
x_scaled = scaler.fit_transform(df[['calories','fat','carbohydrates','protein','cholesterol','sodium'	,'fiber']])
```

```python
x_scaled
```

```python
#combune features
x_combined = np.hstack([x_scaled,X.toarray()])
```

```python
x_combined
```

```python
# Train KNN Model
knn = NearestNeighbors(n_neighbors=3, metric='euclidean')
knn.fit(x_combined)
```

```python
#function to recommend recipe
def recommendation(input_features):
  input_features_scaled = scaler.transform([input_features[:7]])
  input_ingredients_tranformed = vector.transform([input_features[7]])
  input_combined = np.hstack([input_features_scaled,input_ingredients_tranformed.toarray()])
  distances, indices = knn.kneighbors(input_combined)
  recommendations = df.iloc[indices[0]]
  return recommendations[['recipe_name','ingredients_list','image_url']]
```

```python
#example
input_features = [15, 36, 1, 42, 21, 81, 2, 'pork belly, smoked paprika, kosher salt']
recommendations = recommendation(input_features)
recommendations
```
